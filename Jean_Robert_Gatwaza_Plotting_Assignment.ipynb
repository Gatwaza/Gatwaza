```python
# YOUR CODE HERE
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is the DataFrame containing your dataset
plt.figure(figsize=(10, 6))

# Plot the histogram for Carbon Dioxide values
plt.hist(df['co2'], bins=30, color='skyblue', edgecolor='black')  # Adjust the bin size as needed

# Add labels and title
plt.xlabel('Carbon Dioxide Emissions')
plt.ylabel('Frequency')
plt.title('Distribution of Carbon Dioxide Emissions')

# Show the plot
plt.show()
# the histogram represents symmetric distribution
```

```python
# YOUR CODE HERE
# Histogram for Engine Displacement
plt.figure(figsize=(10, 6))
plt.hist(df['displ'], bins=15, color='salmon', edgecolor='black')  # Adjust the bin size as needed
plt.xlabel('Engine Displacement')
plt.ylabel('Frequency')
plt.title('Distribution of Engine Displacement')
plt.show()

# The histogram below depicts negatively skewed distribution
```

```python
# YOUR CODE HERE
# Histogram for Fuel Consumption
plt.figure(figsize=(10, 6))
plt.hist(df['fuelType'], bins=15, color='lightgreen', edgecolor='black')  # Adjust the bin size as needed
plt.xlabel('Fuel Consumption')
plt.ylabel('Frequency')
plt.title('Distribution of Fuel Consumption')
plt.show()

# The histogram below depicts negatively skewed distribution
```

**TO DO:** create a heat map of engine displacement i.e the column "disp"  vs. co2 production.and interpret it (is there correlation?what did you notice?). Engine Displacement is in litres

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset
plt.figure(figsize=(10, 6))

# Create a heatmap
sns.heatmap(data=df[['displ', 'co2']].corr(), annot=True, cmap='coolwarm', fmt=".2f")

# Add labels and title
plt.xlabel('Engine Displacement (litres)')
plt.ylabel('CO2 Emissions')
plt.title('Correlation between Engine Displacement and CO2 Emissions')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the dataset
plt.figure(figsize=(12, 8))

# Create a correlation matrix
corr_matrix = df.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Add labels and title
plt.xlabel('Column Names')
plt.ylabel('Column Names')
plt.title('Correlation Heatmap')

# Rotate the y-axis labels for better visibility
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
